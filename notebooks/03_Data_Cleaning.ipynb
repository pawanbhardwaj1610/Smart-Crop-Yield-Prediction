## if required to remove

```python
crop_df.drop_duplicates(inplace=True)

weather_df.drop_duplicates(inplace=True)

soil_df.drop_duplicates(inplace=True)
```

## Remove Extra Spaces

```python
crop_df["State"] = crop_df["State"].str.strip()

crop_df["Crop"] = crop_df["Crop"].str.strip()

crop_df["Season"] = crop_df["Season"].str.strip()
```

```python
crop_df["State"].unique()
```

## Standardize Text Case

```python
crop_df["State"] = crop_df["State"].str.title()

crop_df["Crop"] = crop_df["Crop"].str.title()

crop_df["Season"] = crop_df["Season"].str.title()
```

## Verify Numeric Columns

```python
crop_df.describe()
```

```python
crop_df = crop_df[crop_df["Yield"]>=0]
```

## Rename Columns (Optional)


```python
crop_df.rename(columns={
    "Crop_Year":"Year"
}, inplace=True)
```

```python
crop_df.columns
```

```python
crop_df.to_csv("../data/processed/crop_yield_cleaned.csv", index=False)

weather_df.to_csv("../data/processed/weather_cleaned.csv", index=False)

soil_df.to_csv("../data/processed/soil_cleaned.csv", index=False)
```
