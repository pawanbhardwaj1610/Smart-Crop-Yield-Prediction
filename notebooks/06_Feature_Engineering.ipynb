```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/final_df.csv")
```

```python
df.info()
```

```python
le = LabelEncoder()

df["Crop"] = le.fit_transform(df["Crop"])
df["State"] = le.fit_transform(df["State"])
df["Season"] = le.fit_transform(df["Season"])
```

## Scale numerical columns

```python
sc = StandardScaler()

num_col = [
    "Area",
    "Production",
    "Annual_Rainfall",
    "Fertilizer",
    "Pesticide",
    "Yield",
    "avg_temp_c",
    "total_rainfall_mm",
    "avg_humidity_percent",
    "N",
    "P",
    "K",
    "pH"
    
]

df[num_col] = sc.fit_transform(df[num_col])
```

## Feature Selection

```python
corr = df.corr(numeric_only = True)

corr["Yield"].sort_values(ascending = False)
```

## Create Derived Features

```python
df["Production_per_area"] = df["Production"] / df["Area"]

df["Rainfall_per_temperature"] = df["Annual_Rainfall"] / df["avg_temp_c"]
```

```python
corr = df.corr
corr
```

```python
df.to_csv(
    "../data/processed/feature_engineered_dataset.csv",
    index=False
)
```
