```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/final_df.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
##plt.figure(figsize=(8,5))
plt.hist(df["Yield"], bins=30)
plt.title("Distribution of Crop Yield")
plt.xlabel("Yield")
plt.ylabel("Frequency")
plt.show()
```

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(10,8))
plt.imshow(numeric_df.corr(), aspect="auto")
plt.colorbar()
plt.xticks(range(len(numeric_df.columns)), numeric_df.columns, rotation=90)
plt.yticks(range(len(numeric_df.columns)), numeric_df.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
yield_by_crop = (
    df.groupby("Crop")["Yield"]
      .mean()
      .sort_values(ascending=False)
)

yield_by_crop.plot(kind="bar", figsize=(12,5))
plt.title("Average Yield by Crop")
plt.show()
```

```python
yield_by_state = (
    df.groupby("State")["Yield"]
      .mean()
      .sort_values(ascending=False)
)

yield_by_state.plot(kind="bar", figsize=(14,6))
plt.title("Average Yield by State")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["total_rainfall_mm"], df["Yield"])
plt.xlabel("Rainfall")
plt.ylabel("Yield")
plt.title("Rainfall vs Yield")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df["Yield"])
plt.title("Yield Outliers")
plt.show()
```
