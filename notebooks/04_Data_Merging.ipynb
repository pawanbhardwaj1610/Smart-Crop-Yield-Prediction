```python
import pandas as pd
```

```python
crop_df = pd.read_csv("../data/processed/crop_yield_cleaned.csv")
weather_df = pd.read_csv("../data/processed/weather_cleaned.csv")
soil_df = pd.read_csv("../data/processed/soil_cleaned.csv")
```

```python
weather_df.rename(columns={"year":"Year"},
                  inplace = True)
weather_df.rename(columns ={"state":"State"},
                  inplace = True)
```

```python
weather_df.columns
```

```python
soil_df.columns
```

```python
soil_df.rename(columns ={"state":"State"}, 
               inplace = True )
```

```python
soil_df.columns
```

# Merge Crop + Weather

```python
crop_weather = pd.merge(crop_df,weather_df,
                        on =["State","Year"],
                        how = "left")
```

# Merge with Soil

```python
final_df = pd.merge(crop_weather,soil_df,
                    on = ["State"],
                    how = "left")
```

```python
crop_states = set(crop_df["State"].unique())
weather_states = set(weather_df["State"].unique())

print("States in Crop but not in Weather:")
print(sorted(crop_states - weather_states))

print("\nStates in Weather but not in Crop:")
print(sorted(weather_states - crop_states))
```

```python
soil_states = set(soil_df["State"].unique())

print("States in Crop but not in Soil:")
print(sorted(crop_states - soil_states))

print("\nStates in Soil but not in Crop:")
print(sorted(soil_states - crop_states))
```

```python
state_mapping = {
    "Jammu And Kashmir" : "Jammu and Kashmir"
}
```

```python
crop_df["State"] = crop_df["State"].replace(state_mapping)
```

```python
crop_weather = pd.merge(
    crop_df,
    weather_df,
    on = ["State","Year"],
    how = "left"
)
```

```python
final_df = pd.merge(
    crop_weather,
    soil_df,
    on = ["State"],
    how = "left"
)
```
